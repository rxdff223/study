### DML
Data manipulation language,用来对数据库中表的数据记录进行更新

数据插入

insert into student(id,name,age) values(4,'fan',31)

student(,,)也可以不写，但是必须一一对应

数据删除

全部删除，但是表结构还是会存在

delete from student where id = 1;

数据更新

update student set name = 'fan' where age = 31;

# DQL
数据查询-基础查询

select id, name, age from student;

select * from student;全部内容



select * from student where age = 31;

数据查询-分组聚合



select gender,avg(age) from student group by gender



数据查询-排序分页

order by
